```python
%matplotlib inline  
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('life_expectancy_data.csv')
```

```python
df_lifeexpectancy_2015 = df[df['year'] == 2015].groupby(['country']).mean()['life_expectancy'].dropna()
```

```python
#df_lifeexpectancy_2015.mean()
df_lifeexpectancy_2015.std()
```

```python
sns.set_context('paper')
sns.set_style('whitegrid')
sns.distplot(df_lifeexpectancy_2015, kde=False, color='g', bins=5)
plt.xticks([50,60,70,80,90])
plt.xlabel('Life Expectancy (years)')
plt.ylabel('Frequency')
plt.title('Global Life Expectancy Distribution in 2015')
plt.savefig('lifeexpectancy_hist.png', dpi=300, bbox_inches='tight')
```

```python
print("     Five number summary \nGlobal Life expectancy in 2015")
quartiles = np.percentile(df_lifeexpectancy_2015, [25, 50, 75])
data_min, data_max = df_lifeexpectancy_2015.min(), df_lifeexpectancy_2015.max()
print('Min: %.3f' % round(data_min,2))
print('Q1: %.3f' % round(quartiles[0],2))
print('Median: %.3f' % round(quartiles[1],2))
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % data_max)
```

```python
sns.set(style='whitegrid')
sns.boxplot(df_lifeexpectancy_2015, linewidth=2.5, palette ='Set2')
sns.swarmplot(df_lifeexpectancy_2015)
plt.title('Boxplot/Swarm Plot - Global Life Expectancy in 2015')
plt.xlabel('Life expectancy(years)')
plt.savefig('boxplot2015.png', dpi=300, bbox_inches='tight')
```

```python
def bmi_status(bmi):
    print('BMI Weight Status')
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 == bmi < 24.9:
        return 'Normal or Healthy Weight'
    elif 25.0 == bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'
```
